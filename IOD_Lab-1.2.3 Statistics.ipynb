# 9. [BONUS] Plot X, Y and Best Fit Line

**Write a function to calculate best fit.**

![alt text](https://i.imgur.com/B43vOV0.jpg)

[Least Squares Regression (Best Line Fit)]: https://www.mathsisfun.com/data/least-squares-regression.html "Least Squares Regression"

```
# Expected Output:
```

````python
"""CODE DELETED. DO NOT TOUCH THIS TEXT. RUNNING THIS CELL WITH SHIFT-ENTER WILL DELETE THE GRAPH"""
````

````python
import numpy as np
import matplotlib.pyplot as plt

# sample points
X = np.array([0, 5, 10, 15, 20])
Y = np.array([0, 7, 10, 13, 20])

# solve for a and b
def best_fit(X, Y):
    x_squared = X*X
    xy = X * Y
    sum_x = np.sum(X)
    sum_y = np.sum(Y)
    sum_x_squared = np.sum(x_squared)
    sum_xy = np.sum(xy)
    N = len(X)

    m = (N * sum_xy - sum_x * sum_y) / (N * sum_x_squared - sum_x ** 2)
    b = (sum_y - m * sum_x) / N

    return m, b

# solution
m, b = best_fit(X, Y)

# Compute y values using y = mx + b
y = m * X + b  

# plot points and fit line
# ANSWER
plt.scatter(X, Y)
plt.plot(X, y)
plt.show()
````

# 13. [BONUS] Calculate Mean Confidence Interval

Check the following methods:

Expected output:

(172.52, 168.26329573448135, 176.77670426551867)

````python
#def mean_confidence_interval(data, confidence=0.95):
# Calculate m, se

def mean_confidence_interval(data, confidence=0.95):
  a = 1.0 * np.array(data)
  n = len(a)
  m, se = np.mean(a), scipy.stats.sem(a)
  h = se * scipy.stats.t.ppf((1 + confidence) / 2., n-1)
  return m, m-h, m+h

data = [170,167,	174,	179,	179,	187,	179,	183,	179 ,156,	163	,156	,187,	156	,
        167	,156	,174	,170 ,183	,179	,174	,179	,170,	159	,187	]
print(mean_confidence_interval(data))
````
