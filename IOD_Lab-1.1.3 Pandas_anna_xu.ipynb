# 4. Visualisation


#### 4.A Plot a histogram of `age`

````python
plt.hist(df["age"])
plt.xlabel("Age")
plt.ylabel("No. of Employees")
plt.title("Age of All Employees")
plt.show()
````

#### 4.B BoxPlot of `age`

````python
plt.boxplot(df["age"])
plt.ylabel("Age")
plt.xlabel("Distribution")
plt.title("Distribution of Age amongst Employees")
plt.show()
````

#### 4.C Create a scatterplot of `age` and `length_of_service`

````python
plt.scatter(df["age"], df["length_of_service"])
plt.xlabel("Age")
plt.ylabel("Length of Service")
plt.title("Age and Length of Service of Employees")
plt.show()
````
